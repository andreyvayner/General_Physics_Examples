First we will import some Python packages that will allow us to solve an ordinary differential equation and plot our results.

```python
import numpy as np #scientific computing, providing powerful tools for working with multi-dimensional arrays and matrices, along with a wide range of mathematical functions
import matplotlib.pyplot as plt #package dealing with plotting
from scipy.integrate import solve_ivp #To solve scientific and mathematical problems. solve_ivp is used to solve ordinary differential equations
#help(solve_ivp)
```

# Let us revist the problem:

![alt text](SHO_FBD.png)

Below we will define the function that includes the differential equations that we want to solve

```python
def sho(t, initial_condition, k, m):
    """Defines the system of equations for simple harmonic motion.
    y[0] is position (x), y[1] is velocity (v).
    """
    x, v = initial_condition
    dxdt = v
    dvdt = - (k / m) * x  # Hooke's law: F = -kx -> a = -kx/m
    return [dxdt, dvdt]
```

The parameters of our problem, including the initial conditions. 

```python
# Parameters
k = 1.0  # Spring constant
m = 1.0  # Mass
x0 = 1.0  # Initial position
v0 = 0.0  # Initial velocity
```

Defining how long we want to run the integration and how "accurately" we want to know the answer

```python
t_span = (0, 20)  # Time span for simulation
t_eval = np.linspace(*t_span, 1000)  # Time points for solution
```

```python
# Solve the differential equation
sol = solve_ivp(sho, t_span, [x0, v0], t_eval=t_eval, args=(k, m))
```

Now lets plot and examine the results:

```python
# Plot the results
plt.figure(figsize=(8, 5)) #Making a figure with a defined size
plt.plot(sol.t, sol.y[0], label='Position (x)') #plotting the distance vs. time solution
plt.plot(sol.t, sol.y[1], label='Velocity (v)', linestyle='dashed') #plotting the velocity vs. time solution
plt.xlabel('Time (s)',fontsize=15) 
plt.ylabel('Amplitude',fontsize=15)
plt.title('Simple Harmonic Oscillator',fontsize=15)
plt.legend(fontsize='large')
plt.grid()
plt.show()
```
